```python
# The laplacian operator including diagonals
kernel = np.array([3, 3])
kernel = [ [-1, -1, -1], [-1, 8, -1], [-1, -1, -1] ]
```

```python
input_placeholder = tf.placeholder(
    dtype=tf.float32, shape=(1, image.shape[0], image.shape[1], 1))
```

```python
with tf.name_scope('convolution'):
    conv_w = tf.constant(kernel, dtype=tf.float32, shape=(3, 3, 1, 1))
    output = tf.nn.conv2d(input=input_placeholder, filter=conv_w, strides=[1, 1, 1, 1], padding='SAME')
    
```

```python
with tf.Session() as sess:
    result = sess.run(output, feed_dict={
            input_placeholder: image[np.newaxis, :, :, np.newaxis]})
```
